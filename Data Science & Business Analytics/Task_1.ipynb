```python
# Import all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(url)
```

```python
data.head(25)
```

```python
# Plot score distribution
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours studied')
plt.ylabel('Percentage scored')
plt.show()
```

From the graph above, we can see that the number of hours a student spends studying is directly proportional to the percentage the student obtains.

### Preparing the data
Here we divide the data into attributes(input) and labels(output).

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

### Training the algorithm


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
print("Training complete.")
```

```python
# Plotting the regression line
line = regressor.coef_*x+regressor.intercept_

# Plotting for test data
plt.scatter(x, y)
plt.plot(x, line)
plt.show()
```

### Making predictions
The algorithm has been trained. Now, we will make some predictions

```python
# Testing data - hours
print(x_test)

# Predicting the score
y_pred = regressor.predict(x_test)
```

```python
# Comparing actual vs predicted data
df = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df
```

### Evaluating the model
This is the final step where we evaluate the performance of our algorithm.

```python
from sklearn import metrics
print('Mean absolute error: ',
     metrics.mean_absolute_error(y_test, y_pred))
```
